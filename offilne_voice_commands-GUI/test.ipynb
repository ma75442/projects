```python
import os
import pandas as pd
import numpy
import shutil
import random
import string
```

```python
com_dict = {1:'abort',2:'activate',3:'centroid',4:'edge',5:'launch',6:'switch',7:'track',8:'zoom'}
```

```python
print('Choose the actual command: 1-abort  2-activate  3-centroid  4-edge  5-launch  6-switch  7-track  8-zoom')
c = input('>>')
```

```python
label = com_dict[int(c)]
```
